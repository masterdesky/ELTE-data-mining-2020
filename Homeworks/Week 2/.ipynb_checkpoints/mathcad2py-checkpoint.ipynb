```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create initial image

```python
N = 100
a = 0.1

sine = np.zeros(N)
noise = np.zeros(N)
for i in range(N):
    # Pure sine
    s = np.sin(a * i)
    # Add noise
    n = np.random.rand() - 0.5
    
    sine[i] = s
    noise[i] = n
    
# Add noise to the pure sine wave
y = sine + noise
```

```python
fig, axes = plt.subplots(figsize=(14,8))

axislegendsize = 15

axes.plot(range(N), sine, label='Pure sine',
          color='tab:blue', lw=3)
axes.plot(range(N), noise, label='Noise',
          color='tab:grey', lw=2, ls='--')
axes.plot(range(N), y, label='Sum of S + N',
          color='tab:red', lw=3)

axes.legend(loc='lower right', fontsize=axislegendsize)
axes.grid(b='On')

plt.show()
```

## Apply masking function

```python
# It could be defined as a scalar too,
# but ok mathcad, here's this retardness
# _mask : list
#   Contains 6 completely equal scalars (1/7)
_mask = [1/7] * 6
```

```python
# Output of convolution
z = np.zeros(N - 6)

# Convolutional step
for i in range(3, N-3):
    
    z_i = 0
    for k in range(-3, 3):
        z_i += _mask[k + 3] * y[i - k]
        
    z[i - 3] = z_i
```

```python
fig, axes = plt.subplots(figsize=(14,8))

axislegendsize = 15

# Pure sine wave
axes.plot(range(N), sine, label='Pure sine',
          color='tab:blue', lw=3)
# Noise
axes.plot(range(N), noise, label='Noise',
          color='tab:grey', lw=2, ls='--')
# Noise applied to the sine wave
axes.plot(range(N), y, label='Sum of S + N',
          color='tab:red', lw=3)
# Masked noisy sine wave
axes.plot(range(3,N-3), z, label='Masked Y',
          color='tab:green', lw=3, ls='--')

axes.legend(loc='lower right', fontsize=axislegendsize)
axes.grid(b='On')

plt.show()
```
